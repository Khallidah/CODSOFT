### Dropping some columns and assigning it to a new variable name - new_df

```python
new_df = df.drop(['PassengerId','Name','Ticket'],axis =1)
new_df
```

### CONVERTING CATEGORICAL FEATURES TO NUMERICAL FEATURES

```python
new_df['Sex'].unique()
```

### Converting the sex column using the map function/method

```python
new_df['Sex'] = new_df['Sex'].map({'male':0,'female':1})
new_df
```

### Converting the cabin column using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
le = LabelEncoder()
new_df['Cabin'] = le.fit_transform(new_df['Cabin'])
new_df.info()
```

```python
new_df
```

### Converting the Embarked column using the get dummies variable

```python
emb_cat = new_df['Embarked'].value_counts()
emb_cat.index
```

```python
most_occuring_embarked = emb_cat.index[0]
new_df['Embarked'] = new_df['Embarked'].fillna(most_occuring_embarked)

emb = pd.get_dummies(new_df['Embarked'],drop_first = False, prefix = 'Embarked').astype("bool")
emb
```

##### Concatenating the getdummy variable assigned to emb with the original dataset new_df 

```python
new_df = pd.concat([new_df,emb],axis = 1 )

### Dropping Embarked since it is no longer needed.

new_df.drop('Embarked',axis = 1, inplace = True)
new_df
```

#### Filling up age and fare columns with the mean.

```python
new_df['Age'] = new_df['Age'].fillna(new_df['Age'].mean())
```

```python
new_df['Fare'] = new_df['Fare'].fillna(new_df['Fare'].mean())
new_df.isnull().sum()
```

```python
new_df.info()
```

### Building model

```python
y = new_df['Survived']
X = new_df.drop('Survived',axis =1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()

### Fitting my model to the train file

logreg.fit(X_train,y_train)
```

```python
pred = logreg.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred) * 100
```

```python
## pickle
import pickle

pickle_out = open('model.pkl', 'wb')
pickle.dump(logreg, pickle_out)
pickle_out.close()
```
